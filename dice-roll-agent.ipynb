### Dice roll - a simple agentic ai solution (no tool calling in this scenario)

```python
import random
from openai import OpenAI

client = OpenAI()

def roll_dice():
    return random.randint(1,6)

```

```python
target = 6
max_iterations = 20
```

```python
history = ""
iteration = 0

while iteration < max_iterations:
    iteration += 1

    instructions = f"""
    You are playing a dice game. You need to roll a {target} on the dice.

    You have made the following rolls so far:
    {history if history else "No rolls yet"}

    Based on what you have rolled so far, decide your next action:

    Option A: if you have achieved the target, respond with exactly: DONE: <message describing how >
    Option B: If you have not achieved the target, respond exactly: ROLL THE DICE
    """


    response = client.chat.completions.create(

        model = "gpt-4.1-mini",
        messages = [ {"role":"user", "content": instructions}]
    )

    result = response.choices[0].message.content

    # process the result to see if the agent has decided to roll again or not
    if "DONE:" in result:
        print(result)
        break

    # update the history with the new result and print the decision
    if "ROLL THE DICE" in result:
        roll = roll_dice()
        print(f"Rolled: {roll}\n")
        history += f"Roll {iteration}: {roll}\n"
        
# stop the loop if the agent has decided to stop or if max_iterations is reached
print(f"Game ended after {iteration} iterations")
    
```
